```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale,StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score,mean_squared_error
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
import warnings
warnings.filterwarnings("ignore",category=DeprecationWarning)
warnings.filterwarnings("ignore",category=FutureWarning)
```

```python
df = pd.read_csv("diabetes.csv")
y = df["Outcome"]
X = df.drop(["Outcome"],axis=1)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
rfc_model = RandomForestClassifier().fit(X_train,y_train)
```

```python
pred_y = rfc_model.predict(X_test)
accuracy_score(y_test,pred_y)
```

# model tuning

```python
rfc_model = RandomForestClassifier()
```

```python
rfc_params = {"max_depth":[None,2,3,4],
             "min_samples_split":[2,3,4]}
```

```python
rfc_cv_model = GridSearchCV(rfc_model,rfc_params,cv=5,n_jobs=-1,verbose=2).fit(X_train,y_train)
```

```python
rfc_cv_model.best_params_
```

```python
rfc_tuned = RandomForestClassifier(max_depth=3,min_samples_split=3).fit(X_train,y_train)
```

```python
pred_y = rfc_tuned.predict(X_test)
accuracy_score(y_test,pred_y)
```

## importance

```python
feature_imp = pd.Series(rfc_tuned.feature_importances_,index=X_train.columns).sort_values(ascending=False)
ax =sns.barplot(x=feature_imp,y=feature_imp.index,palette="Reds")
plt.xlabel("Değişken önem skorları")
plt.ylabel("Değişkenler")
plt.title("Önem düzeyi grafiği")
plt.show()

```
